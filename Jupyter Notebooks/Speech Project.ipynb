```python
import python_speech_features
import os
import librosa
import random
import numpy as np
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras import layers, models
```

```python
dataset_path = "/home/main_dude/Projects/speech_tech/new_dataset/spk_independent"
targets = [name for name in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, name))]
targets.sort()
print(targets)
```

# Extracting MFCCs

```python
speech_files = []
y = []
for index, target in enumerate(targets):
    speech_files.append(os.listdir(os.path.join(dataset_path, target)))
    y.append(np.ones(len(speech_files[index])) * index)
speech_files = [item for sublist in speech_files for item in sublist]
y = [int(item) for sublist in y for item in sublist]
```

```python
for index, target in zip(y, speech_files):
    speech_path = os.path.join(dataset_path, str(index), target)
    print(speech_path)
```

```python
mfcc_features = []
mfcc_delta = []
mfcc_ddelta = []
y_new = []
invalid_count = 0
for index, target in zip(y, speech_files):
    speech_path = os.path.join(dataset_path, str(index), target)
    signal, fs = librosa.load(speech_path, sr = 16000)
    mfccs = python_speech_features.base.mfcc(signal, 
                                             samplerate = fs,
                                             winlen = 0.025,
                                             winstep = 0.01,
                                             numcep = 13,
                                             nfilt = 20,
                                             nfft = 2048,
                                             lowfreq=100, 
                                             highfreq=None,
                                             preemph = 0.97,
                                             ceplifter = 0,
                                             appendEnergy = True,
                                             winfunc = np.hanning)
    delta = librosa.feature.delta(mfccs, mfccs.shape[1])
    ddelta = librosa.feature.delta(delta, delta.shape[1])
    if mfccs.shape[0] != 99:
        invalid_count += 1
    else:
        y_new.append(index)
        mfcc_delta.append(delta) 
        mfcc_features.append(mfccs)  
        mfcc_ddelta.append(ddelta)
```

```python
print(invalid_count)
```

```python
y = y_new
mfcc_features= np.array(mfcc_features)
mfcc_delta = np.array(mfcc_delta)
mfcc_ddelta = np.array(mfcc_ddelta)
mfcc_26 = []
for i in range(mfcc_features.shape[0]):
    mfcc_26_temp = np.hstack((mfcc_features[i], mfcc_delta[i]))
    mfcc_26_temp = np.hstack((mfcc_26_temp, mfcc_ddelta[i]))
    mfcc_26.append(mfcc_26_temp.transpose())
print(mfcc_26[0].shape)
```

```python
fig = plt.figure()
plt.imshow(mfcc_26[0], cmap='inferno', origin='lower')
plt.xlabel('Time (ms)')
plt.ylabel('MFCC Co-efficients')
plt.show()
```

```python
x_train, x_val_test, y_train, y_val_train = train_test_split(mfcc_26, y, test_size = 0.2, random_state = 69)
x_val, x_test, y_val, y_test = train_test_split(x_val_test, y_val_train, test_size = 0.5, random_state = 69)
x_train = np.array(x_train)
x_val = np.array(x_val)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_val = np.array(y_val)
y_test = np.array(y_test)
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
print(y_train.shape)
print(y_val.shape)
print(y_test.shape)
```

# MODEL TRAINING

```python
x_train = x_train.reshape(x_train.shape[0], 
                          x_train.shape[1], 
                          x_train.shape[2], 
                          1)
x_val = x_val.reshape(x_val.shape[0], 
                      x_val.shape[1], 
                      x_val.shape[2], 
                      1)
x_test = x_test.reshape(x_test.shape[0], 
                        x_test.shape[1], 
                        x_test.shape[2], 
                        1)
sample_shape = x_test.shape[1:]
print(x_train.shape)
print(sample_shape)
```

```python
keras.backend.clear_session()
model = models.Sequential()
model.add(layers.Conv2D(32, (2, 2), activation='relu', input_shape=sample_shape))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Conv2D(32, (2, 2), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Dropout(0.5))
model.add(layers.Conv2D(64, (2, 2), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(len(targets), activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

```python
history = model.fit(x_train, 
                    y_train, 
                    epochs=50, 
                    batch_size=30, 
                    validation_data=(x_val, y_val))
```

```python
metrics = history.history
plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()
```

```python
models.save_model(model, "digit_recog.h5")
```

# Model Testing

```python
for i in range(10, 15):
    print('Answer:', y_test[i], ' Prediction:', model.predict(np.expand_dims(x_test[i], 0)))
```

```python
model.evaluate(x=x_test, y=y_test)
```

```python
y_pred = np.argmax(model.predict(x_test), axis=1)
y_true = y_test

test_acc = sum(y_pred == y_true) / len(y_true)
print(f'Test set accuracy: {test_acc:.0%}')
```
